```python
# 💖 Cute Interactive Electronics Lab 💖
import numpy as np
import plotly.graph_objects as go
from ipywidgets import FloatSlider, Dropdown, interactive
from IPython.display import display

# --- Create sliders and dropdown ---
voltage_slider = FloatSlider(
    value=3.3, min=0, max=5000.0, step=0.1, description='Voltage (V):',
    style={'description_width':'initial'}
)
resistance_slider = FloatSlider(
    value=1000, min=1, max=10000, step=1, description='Resistance (Ω):',
    style={'description_width':'initial'}
)
voltage_type_dropdown = Dropdown(
    options=['constant', 'ramp', 'AC'],
    value='constant',
    description='Voltage Type:',
    style={'description_width':'initial'}
)

# --- Function to update plot ---
def update_plot(voltage, resistance, voltage_type):
    time = np.linspace(0, 1, 100)
    
    # Compute current based on voltage type
    if voltage_type == 'constant':
        current_wave = np.ones_like(time) * (voltage / resistance)
    elif voltage_type == 'ramp':
        voltage_wave = np.linspace(0, voltage, len(time))
        current_wave = voltage_wave / resistance
    elif voltage_type == 'AC':
        voltage_wave = voltage * np.sin(2 * np.pi * 5 * time)  # 5 Hz
        current_wave = voltage_wave / resistance
    
    # Create Plotly figure
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=time,
        y=current_wave,
        mode='lines+markers',
        marker=dict(symbol='circle', size=10, color='pink'),
        line=dict(color='hotpink', width=4),
        name='Current'
    ))
    
    fig.update_layout(
        title=f"Current through resistor: {current_wave[-1]:.3f} A 💖",
        xaxis_title="Time (s)",
        yaxis_title="Current (A)",
        plot_bgcolor='mistyrose',
        paper_bgcolor='lavenderblush',
        font=dict(color='deeppink', family="Comic Sans MS"),
        height=400
    )
    
    fig.show()

# --- Create interactive widget ---
interactive_plot = interactive(
    update_plot,
    voltage=voltage_slider,
    resistance=resistance_slider,
    voltage_type=voltage_type_dropdown
)

display(interactive_plot)
```
