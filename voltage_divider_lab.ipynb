```python
# voltage_divider_lab.ipynb
import numpy as np
import plotly.graph_objects as go
from ipywidgets import FloatSlider, interactive, VBox
from IPython.display import display

# --- Interactive sliders ---
V_in_slider = FloatSlider(value=5, min=0, max=20, step=0.1, description='Input Voltage (V):')
R1_slider = FloatSlider(value=1000, min=1, max=10000, step=1, description='R1 (Ω):')
R2_slider = FloatSlider(value=2000, min=1, max=10000, step=1, description='R2 (Ω):')

display(VBox([V_in_slider, R1_slider, R2_slider]))

# --- Update function ---
def update_plot(V_in, R1, R2):
    V_out = V_in * R2 / (R1 + R2)
    I_total = V_in / (R1 + R2)
    
    t = np.linspace(0, 1, 100)
    V_mid_wave = V_out * np.sin(np.pi * t)
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=t,
        y=V_mid_wave,
        mode='lines+markers',
        line=dict(color='hotpink', width=4),
        marker=dict(color='deeppink', size=8, symbol='circle'),
        name='Midpoint Voltage'
    ))
    
    fig.update_layout(
        title=f"Voltage at midpoint: {V_out:.2f} V 💖",
        xaxis_title="Time (s)",
        yaxis_title="Voltage (V)",
        plot_bgcolor='mistyrose',
        paper_bgcolor='lavenderblush',
        font=dict(color='deeppink', family='Comic Sans MS'),
        height=400
    )
    
    fig.show()

interactive_plot = interactive(update_plot, V_in=V_in_slider, R1=R1_slider, R2=R2_slider)
display(interactive_plot)
```
