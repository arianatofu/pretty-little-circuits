```python
import numpy as np
import plotly.graph_objects as go
from ipywidgets import interactive, FloatSlider, Dropdown
from IPython.display import display as ipydisplay, Markdown

display(Markdown("### Lab 1: Capacitor Charging & Discharging"))

# Create sliders with precise decimal readout
V_s_slider = FloatSlider(value=5, min=0, max=20, step=0.1, description='Supply Voltage (V):')
R_slider   = FloatSlider(value=1000, min=1, max=10000, step=1, description='Resistance (Ω):')
C_slider   = FloatSlider(value=0.001, min=0.000001, max=0.01, step=0.000001,
                         description='Capacitance (F):', readout_format='.6f')
mode_dropdown = Dropdown(options=['charge','discharge'], value='charge', description='Mode:')

def update_lab1(V_s, R, C, mode):
    t = np.linspace(0, 5*R*C, 100)
    V_c = V_s * (1 - np.exp(-t/(R*C))) if mode == 'charge' else V_s * np.exp(-t/(R*C))
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=t, y=V_c,
        mode='lines+markers',
        line=dict(color='hotpink', width=4),
        marker=dict(color='deeppink', size=8)
    ))
    
    fig.update_layout(
        title=f"Capacitor {mode.capitalize()} 💖",
        xaxis_title="Time (s)",
        yaxis_title="Voltage (V)",
        yaxis=dict(tickformat=".6f"),  # Show 6 decimals on axis
        plot_bgcolor='mistyrose',
        paper_bgcolor='lavenderblush',
        font=dict(color='deeppink', family='Comic Sans MS'),
        height=400
    )
    
    ipydisplay(fig)

interactive_lab1 = interactive(update_lab1, V_s=V_s_slider, R=R_slider, C=C_slider, mode=mode_dropdown)
interactive_lab1
```

```python
display(Markdown("### Lab 2: RC Low-Pass Filter"))

f_slider   = FloatSlider(value=5, min=0.1, max=50, step=0.1, description='Frequency (Hz):')
V_in_slider = FloatSlider(value=5, min=0, max=20, step=0.1, description='Input Voltage (V):')
R_slider2   = FloatSlider(value=1000, min=1, max=10000, step=1, description='Resistance (Ω):')
C_slider2   = FloatSlider(value=0.001, min=0.000001, max=0.01, step=0.000001,
                          description='Capacitance (F):', readout_format='.6f')

def update_lab2(f, V_in, R, C):
    t = np.linspace(0, 1, 200)
    Vin_wave = V_in * np.sin(2 * np.pi * f * t)
    
    tau = R*C
    Vout_wave = np.zeros_like(t)
    Vout = 0
    dt = t[1]-t[0]
    for i in range(len(t)):
        Vout += (dt/tau)*(Vin_wave[i]-Vout)
        Vout_wave[i] = Vout
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=t, y=Vin_wave, mode='lines', line=dict(color='pink', width=3), name='Input Voltage'))
    fig.add_trace(go.Scatter(x=t, y=Vout_wave, mode='lines+markers', line=dict(color='deeppink', width=4), marker=dict(size=6), name='Capacitor Voltage'))
    
    fig.update_layout(
        title=f"RC Low-Pass Filter 💖",
        xaxis_title="Time (s)",
        yaxis_title="Voltage (V)",
        yaxis=dict(tickformat=".6f"),
        plot_bgcolor='mistyrose',
        paper_bgcolor='lavenderblush',
        font=dict(color='deeppink', family='Comic Sans MS'),
        height=400
    )
    
    ipydisplay(fig)

interactive_lab2 = interactive(update_lab2, f=f_slider, V_in=V_in_slider, R=R_slider2, C=C_slider2)
interactive_lab2
```

```python
display(Markdown("### Lab 3: Capacitors in Series & Parallel"))

C1_slider = FloatSlider(value=0.001, min=0.000001, max=0.01, step=0.000001,
                        description='C1 (F):', readout_format='.6f')
C2_slider = FloatSlider(value=0.001, min=0.000001, max=0.02, step=0.000001,
                        description='C2 (F):', readout_format='.6f')
V_total_slider = FloatSlider(value=5, min=0, max=20, step=0.1, description='Total Voltage (V):')
mode_dropdown3 = Dropdown(options=['series','parallel'], value='series', description='Mode:', )

def update_lab3(C1, C2, mode, V_total):
    C_eq = 1/(1/C1 + 1/C2) if mode=='series' else C1 + C2
    Q = C_eq * V_total
    V1 = Q / C1
    V2 = Q / C2
    
    fig = go.Figure()
    fig.add_trace(go.Bar(
        x=['C1','C2'],
        y=[V1,V2],
        marker_color=['deeppink','hotpink'],
        text=[f'{V1:.6f} V', f'{V2:.6f} V'],
        textposition='auto'
    ))
    
    fig.update_layout(
        title=f"Capacitors in {mode.capitalize()} 💖 (Total Voltage: {V_total} V)",
        yaxis_title="Voltage (V)",
        yaxis=dict(tickformat=".6f"),
        plot_bgcolor='mistyrose',
        paper_bgcolor='lavenderblush',
        font=dict(color='deeppink', family='Comic Sans MS'),
        height=400
    )
    
    ipydisplay(fig)

interactive_lab3 = interactive(update_lab3, C1=C1_slider, C2=C2_slider, mode=mode_dropdown3, V_total=V_total_slider)
interactive_lab3
```
